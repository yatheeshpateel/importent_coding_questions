# 1. Write a code to reverse a number

```python
number = int(input("enter the number "))
reverse =0
while number>0:
    remainder = number%10
    reverse = reverse*10+remainder
    number = number//10
print("the reverse of a number is",reverse)
```

# 3. Write the code to find the Fibonacci series upto the nth term.

```python
n = int(input('enter a element'))
f1= 0
f2= 1
print(f1,end=' ')
print(f2,end=' ')
f3 = f1+f2
while(f3<=n):
    print(f3,end=' ')
    f1=f2
    f2=f3
    f3 = f1+f2
```

# 4  Write code of Greatest Common Divisor 

```python
def computegcd(a,b):
    if b == 0:
        return a
    else:
        return computegcd(b,a%b)
num1 = int(input("enter the number:"))
num2 = int(input("enter the second number:"))
computegcd(num1,num2)
```

# 5. Write code of  Perfect number


```python
n = int(input("enter the number"))
sum = 0

for i in range(1,n):
    if n%i == 0:
        sum = sum+i
if sum==n:
    print(n,"is a perfect number")
else:
    print(n,"is not a prime number")
```

#  6. Write code to Check if two strings are Anagram or not

```python
#take user input
String1 = input("enter the str1")
String2 = input("enter the str2")
#check if length matches
if len(String1) != len(String2):
    #if False
    print("not anagram")
else:
    #sorted function sort string by characters
    String1 = sorted(String1)
    String2 = sorted(String2)
#check if now strings matches
if String1 == String2:
#if True
    print("anagram")
else:
    print("not anagram")
```

#  6. Write code Check if the given string is Palindrome or not

```python
#take user input
String1 = input('Enter the String :')
#initialize string and save reverse of 1st string
String2 = String1[::-1]
#check if both matches
if String1 == String2:
    print('String is palindromic')
else:
    print('Strings is not palindromic')
```
